Function used to determine whether the product is seasonal or not based on Jaccard score

```python
def isSeasonal(product):
  max_score = 0.0

  for index in range(seasonal_product.shape[0]):
    score = 0.0
    candidate = seasonal_product.iloc[index]
    candidate = candidate[:-1] # Remove the cluster value
    candidate = candidate[1:] # Remove the product id
    candidate = np.array(candidate)
    score = jaccard_score(candidate, product)
    
    if score > max_score:
      max_score = score
    
  return max_score
```

Creation of a seasonal product

```python
new_product = seasonal_product.iloc[1]
new_product = new_product[:-1]
new_product = new_product[1:]
new_product = np.array(new_product)
```

```python
new_product_2 = np.ones(60)
```

```python
score_1 = isSeasonal(new_product)
score_2 = isSeasonal(new_product_2)
```

The threshold is set to 70%. The product time series is seasonal if it is  similar to at least 70% of at least one product in the Seasonal products table

```python
if score_1 > 0.7:
  print('The new product is a seasonal product')
else:
  print('The new product is not a seasonal product')
```
